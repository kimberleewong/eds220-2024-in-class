## Selecting a single column

Simples case: select a single column by column name

General syntax:
```python
df['column_name']
```

This is an example of **label-based subsetting**, which means we want to select data from our df using the *names* of the columns, not their position.

### Example
Select number of bird species observed at the Mugu lagoon in spring.

````python
# Select a single column by using square brackets []
mul_spring = df["MUL_spring"]

mul_spring
````

````python
# Confirm the type of the output
print(type(mul_spring))
````

Recall: we can think of a `pandas.Dataframe` as a dictionary of its columns.
Then, we can access a single column using the name as the *key*, just like we would in a dictionary.
`df['column_name']`

We can also do label-based subsetting of a single column using attrribute syntax:
```python
df.column_name
```
Example:

````python
df.MUL_spring
````

**Favor `df['column_name']` instead of df.column_name**

Why?

- `df['column_name']` avoids conflicts with `pandas.Dataframes` methods and attributes. For example, of the `df` has a column named `count`, its amibiguous wheter `pd.count` is referring to the `count()` method or the  `count` column.
- `df['column_name']` can take any column name
- `df.column` only workds if the column name has no spaces or special characters

## Selecting multiple columns...

### ... using a list of column names

Seelct multiple columns in a single cell by passing a list of column names to square brackets `[]`

```python
df[['col1', 'col2', 'col3']]
```

Notice the "double square brackets". This is becasuse we are passing the list of names ``['col1', 'col2', 'col3']` to the selection brackets `[]`.

### ... using a slice

To select a slice of the columsn we will use a special case of **`loc` selection**. General syntax:
```python
df.loc[ :, 'column_start': 'column_end']
```
`column_start` and `column_end` = the starting and endng points of the column slice we want to subset from the data frame.

Notice:

- the first value passed to `loc` is used for selecting rows, using a colon ':' as the row-selection parameter means "select all the rows"

- the slice of the resulting data frame includes both ends of the slice.

````python
# Select columns between 'CSM_winter' and 'MUL_fall'

csm_mul = df.loc[:, 'CSM_winter': 'MUL_fall']
csm_mul.head()
````

## Selecting rows...

### ... using a condition
To select rows that satisfy a particular condition the general syntax is
```python
df[condition_on_rows]
```

THe `condition_on_rows` can be many things, lets see some usual scenarios.

### Example
We are interested in data after 2020.

````python
# Select all rows with years > 2020
post_2020 = df[df['year'] > 2020]
post_2020
````

condition for our rows = `df['year'] > 2020`
this is a `pandas.Series` with boooolean values (`True` or `False`) indicating which rows satisfy the condition year > 2020

````python
# Check the type of df['year'] > 2020
print(type(df['year'] > 2020))

# Print the boolean series
df['year'] > 2020
````

When we pass such a series of boolean values to the selection breackets `[]` we keep only the rows that correspond to a `True` value.

## Check in

Get the subset of the data frame on which the San Dieguito Wetland has at least 75 species recorded during spring

````python
sdw_75 = df[df['SDW_spring'] >= 75]
sdw_75
````

#### Example

We want to look at data from years 2012 to 2015 including both years
We can use this `between` operator method in our condition.

````python
subset = df[df['year'].between(2012,2015)]
subset
````

Let's break it down:

1. df['year'] = column with the year values, a `panda.Series`
2. in `df['year'].between` we are accessing a method for the `panda.Series` using '.'
3. (2012,2015) = parameters, this includes both endpoints
4. `df['year'].between(2012,2015)` = `panda.Series` of boolean values indicating which rows have year value between 2012 ans 2015
5. when we put `df['year'].between(2012,2015)` inside the selection brackets `[]` we obtain the rows we wanted to select

## Avoid using loc for selecting rows

It is equivalent to write the following two:

````python
# Select rows with the year <2015
df[df['year']<2015]
````

````python
# Select rows with the year <2015 using loc
df.loc[df['year']<2015, :]
````

We prefer first syntax because it's simpler.

### using multiple conditions

We can combine multiple conditions to select rows by surrounding each condition in parenthese `()` and using the or operator `|` and the and operator `&`.

## Example: or
Select rows in which CSM has more than 50 species registered in winter *or* fall:

````python
df[(df['CSM_winter']>50) | (df['CSM_fall']>50)] 
````

## Example: and
Select rows in which both CSM and SDW have more than 60 species in spring

````python
df[(df['CSM_spring']>60) & (df['SDW_spring']>60)]
````

empty dataframe! no rows satisfy the conditions

## Select rows by position

Sometmes we may want to selet certain rows based on their *actual position* in the data frame. In other words, using **position-based subsettings**.

To do this, we use **iloc selection** syntax.

```python
df.iloc[row-indices]
```

`iloc` = stands for integer-location based indexing

## Example:

````python
# select the fifth row (index = 4)
df.iloc[4]
````

````python
# Select rows 10 - 13
df.iloc[9:14]
````

### Selecting rows and columns simultaneously

Can be done using `loc` (labels) or `iloc` (positions).

### ... by labels or conditions
When we want to select rows and columns simultaneaously by labels (including using condiitons) we can use `loc` selection with the syntax:

```python
df.loc[row-selection, column-selection]
```
Example:

Select winter surveys from MUL and TJE after 2020

````python
df.loc[df['year'] > 2020 , ['MUL_winter', 'TJE_winter']]
````

## Selecting by position

We want rows 3 - 7 (including 7) and columns 4th and 5th column

````python
df.iloc[3:8, [3,4]]
````
